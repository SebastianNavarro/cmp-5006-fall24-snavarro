```python
# Suppose that APNDJI or XYGROBO are ciphertexts that are obtained from encryption using the Shift Cipher. Show in each case that there are two ”meaningful” plaintexts that could encrypt to the given ciphertext. 

# Shift Cipher
def shift_cipher_forward(text, shift):
    result = ""
    for i in range(len(text)):
        char = text[i]
        if char.isupper():
            result += chr((ord(char) + shift - 65) % 26 + 65)
        else:
            result += chr((ord(char) + shift - 97) % 26 + 97)
    return result

def shift_cipher_backward(text, shift):
    result = ""
    for i in range(len(text)):
        char = text[i]
        if char.isupper():
            result += chr((ord(char) - shift - 65) % 26 + 65)
        else:
            result += chr((ord(char) - shift - 97) % 26 + 97)
    return result

def brute_force(text):
    print("Forward:")
    for i in range(1, 26):
        print(f"Shift {i}: {shift_cipher_forward(text, i)}")
    print("\nBackward:")
    for i in range(1, 26):
        print(f"Shift {i}: {shift_cipher_backward(text, i)}")

cypher_text = "XYGROBO"
brute_force(cypher_text)
```

```python
def find_divisors(n):
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)
    return divisors

def find_common_divisor(numbers):
    common_divisors = find_divisors(numbers[0])
    for i in range(1, len(numbers)):
        common_divisors = list(set(common_divisors) & set(find_divisors(numbers[i])))
    return common_divisors

def find_repeated_words(text, sequence_length):
    sequences = {}
    
    # Iterar a través del texto y extraer secuencias
    for i in range(len(text) - sequence_length + 1):
        sequence = text[i:i + sequence_length]
        
        if sequence in sequences:
            sequences[sequence].append(i)
        else:
            sequences[sequence] = [i]

    # Filtrar solo las secuencias que aparecen más de una vez
    repeated_sequences = {seq: pos for seq, pos in sequences.items() if len(pos) > 1}

    return repeated_sequences

def split_into_columns(text, num_columns):
    columns = {f"Columna {i + 1}": [] for i in range(num_columns)}

    # Iterar a través del texto y asignar caracteres a las columnas
    for index, char in enumerate(text):
        column_index = index % num_columns
        columns[f"Columna {column_index + 1}"].append(char)

    # Formatear el resultado como cadenas
    formatted_columns = {key: ' '.join(value) for key, value in columns.items()}

    return formatted_columns

# Contar cuántas veces aparece cada letra en cada columna y hacer un match solo con la letra E
def count_letters_and_match(text):
    letter_count = {}
    for char in text:
        if char.isalpha():
            if char in letter_count:
                letter_count[char] += 1
            else:
                letter_count[char] = 1

    # Ordenar las letras por frecuencia (de mayor a menor)
    sorted_letters = sorted(letter_count.items(), key=lambda x: x[1], reverse=True)

    # Hacer match solo con la letra E
    matches = {}
    for letter, _ in sorted_letters:
        matches[letter] = 'E'  # Asignar la letra 'E' para el match

    return letter_count, matches

# Función para calcular la distancia entre letras
def calculate_distance(letter, matched_letter):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    letter_index = alphabet.index(letter)
    matched_index = alphabet.index(matched_letter)
    distance = (matched_index - letter_index) % 26  # Distancia circular
    return distance

def getKeyLetter(letter, distance):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    letter_index = alphabet.index(letter)
    matched_index = (letter_index - distance) % 26  # Restamos la distancia para encontrar la clave
    matched_letter = alphabet[matched_index]
    return matched_letter

# Ejemplo de uso
text = "KCCPKBGUFDPHQTYAVINRRTMVGRKDNBVFDETDGILTXRGUDDKOTFMBPVGEGLTGCKQRACQCWDNAWCRXIZAKFTLEWRPTYCQKYVXCHKFTPONCQQRHJVAJUWETMCMSPKQDYHJVDAHCTRLSVSKCGCZQQDZXGSFRLSWCWSJTBHAFSIASPRJAHKJRJUMVGKMITZHFPDISPZLVLGWTFPLKKEBDPGCEBSHCTJRWXBAFSPEZQNRWXCVYCGAONWDDKACKAWBBIKFTIOVKCGGHJVLNHIFFSQESVYCLACNVRWBBIREPBBVFEXOSCDYGZWPFDTKFQIYCWHJVLNHIQIBTKHJVNPIST"
repeated_sequences = find_repeated_words(text, 6)

distancias = []

print(f"Texto: {text}")

# Imprimir resultados
print("Secuencias repetidas y sus posiciones:")
for seq, positions in repeated_sequences.items():
    print(f"\nSecuencia: '{seq}' encontrada en posiciones: {positions} con distancia: {positions[1] - positions[0]}")
    print(f"Divisores de la distancia: {find_divisors(positions[1] - positions[0])}")
    distancias.append(positions[1] - positions[0])
    
print(f"\nDistancias: {distancias}")
print(f"Divisores comunes de las distancias: {find_common_divisor(distancias)}")

# Dividir el texto en columnas
num_columns = 6
columns = split_into_columns(text, num_columns)

# Lista para almacenar las primeras letras de clave de cada columna
first_key_letters = []

# Imprimir columnas
print("\nTexto dividido en columnas:")
for key, value in columns.items():
    print(f"\n{key}: {value}")
    letter_count, matches = count_letters_and_match(value)
    
    # Ordenar el conteo de letras de mayor a menor
    sorted_letter_count = dict(sorted(letter_count.items(), key=lambda x: x[1], reverse=True))
    
    print(f"Conteo de letras (de mayor a menor): {sorted_letter_count}")

    # Calcular distancias y mostrar las letras de la clave
    print(f"Letras de la clave y sus distancias:")
    i = 0
    for letter, matched_letter in matches.items():
        
        distance = calculate_distance(letter, matched_letter)
        key_letter = getKeyLetter(letter, distance)
        print(f"{letter} -> {matched_letter}, Distancia: {distance}, Letra de la clave: {key_letter}")
        
        if i == 0:
            first_key_letters.append(key_letter)
            i = 1
        
        
# Imprimir las primeras letras de clave de todas las columnas
print(f"\nPrimeras letras de clave de todas las columnas: {' '.join(first_key_letters)}")
```
